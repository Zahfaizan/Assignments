# **Here we calculate the mean, median, mode, and standard deviation for these columns**

**Mean**

```python
print('volume mean :',df['Volume'].mean())
print('average price mean :',df['Avg Price'].mean())
print('total sales value mean :',df['Total Sales Value'].mean())
print('discount rate mean :',df['Discount Rate (%)'].mean())
print('discount amount mean :',df['Discount Amount'].mean())
print('Net sales value mean:',df['Net Sales Value'].mean())
```

 **Median**

```python
print('volume median :',df['Volume'].median())
print('average price median :',df['Avg Price'].median())
print('total sales value median :',df['Total Sales Value'].median())
print('discount rate median :',df['Discount Rate (%)'].median())
print('discount amount median:',df['Discount Amount'].median())
print('Net sales value  median:',df['Net Sales Value'].median())
```

**Mode**

```python
print('volume mode :',df['Volume'].mode())
print('average price mode :',df['Avg Price'].mode())
print('total sales value mode :',df['Total Sales Value'].mode())
print('discount rate mode :',df['Discount Rate (%)'].mode())
print('discount amount mode:',df['Discount Amount'].mode())
print('Net sales value  mode:',df['Net Sales Value'].mode())
```

**Standard Deviation**

```python
print('volume standard deviation  :',df['Volume'].std())
print('average price standard deviation  :',df['Avg Price'].std())
print('total sales value standard deviation  :',df['Total Sales Value'].std())
print('discount rate standard deviation  :',df['Discount Rate (%)'].std())
print('discount amount standard deviation :',df['Discount Amount'].std())
print('Net sales value  standard deviation :',df['Net Sales Value'].std())
```

From  the dataset given i have  observed that there are no null values.  I have calculated the mean, median, mode, and standard deviation for the numerical columns. However, when it comes to the mode, we found multiple modes with the same frequency or cases where all values are unique.

**Histogram**

```python
for col in df_numeric:
    plt.figure(figsize=(5,5))
    plt.hist(df_numeric[col], bins=5, color='blue', edgecolor='black')
    plt.title(f'Histogram of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.grid(False)
    plt.show()
```

# **Analyzing the distribution**

**Skewness**

```python
for col in df_numeric:
  skewness=round(df[col].skew(),2)
  print(f'Skewness of {col}:{skewness}')
```

```python
df_numeric.info()
```

**Outliers**

```python
for col in df_numeric:
  Outliers=df[(df[col] > df[col].mean() + 3*df[col].std()) | (df[col]<df[col].mean()-3*df[col].std())]
  print(f'Outliers of {col}:{Outliers}')
```

#  caluculate the Z-Score by using the mean and std values in the data set

```python
for col in df_numeric:
  mean=df[col].mean()
  std=df[col].std()
  df_standard[col]=(df[col]-mean)/std
```

```python
df_standard
```

```python
df_standard.head()
```

# **Comparing the Before and After the  standardization**

# **Befor standardization**

```python
plt.figure(figsize=(5,5))
plt.subplot(1,2,1)
sns.histplot(df[col],kde=False,color='blue')
plt.title(f'Before standardizaion:{col}')
plt.xlabel(col)
plt.show()
```

#After Standardization

```python
plt.figure(figsize=(5,5))
plt.subplot(1,2,2)
sns.histplot(df_standard[col],kde=False,color='purple')
plt.title(f'After standardization:{col}')
plt.xlabel(col)
plt.show()
```
